```python
import numpy as np
import matplotlib.pyplot as plt
```

Se definen los valores de $\alpha$ y $\beta$ junto a los coeficientes:

```python
n1 = 1
n2 = 1.5

beta = n2/n1

def alpha (x):
    num = np.sqrt(1 - ((n1/n2)*np.sin(x))**2)
    den = np.cos(x)
    return num/den

def R (x):
    return ((1-(alpha(x)*beta))/(1 + (alpha(x) * beta)))**2

def T (x):
    return alpha(x)*beta* (2/(1 + (alpha(x) * beta)))**2
```

Se realizan las gr√°ficas:

```python
x = np.linspace(0, np.pi/2, 100)
y1 = R(x)
y2 = T(x)
y3 = R(x) + T(x)

plt.plot(x, y1, label = "R")
plt.plot(x, y2, label = "T", color = "g")
plt.plot(x, y3, label = "T + R", color = "r")
plt.title('Coeficientes de Reflexi√≥n y Transmisi√≥n')
plt.xlabel('√Ångulo')
plt.ylabel('Valor de los coeficientes')
plt.grid(True)
plt.legend()
plt.show()
```
